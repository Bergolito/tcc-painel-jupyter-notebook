```python
import numpy as np
import matplotlib.pyplot as plt

# Criar um array 9x9 com zeros
array = np.zeros((9, 9), dtype=int)

# Coluna a ser preenchida (começando em 0)
coluna_a_preencher = 3

# Preencher a coluna com números de 1 a 9
array[:, coluna_a_preencher] = np.arange(1, 10)

# Criar uma máscara para destacar as bordas
mask = np.zeros_like(array)
mask[0, :] = 1  # Primeira linha
mask[-1, :] = 1  # Última linha
mask[:, 0] = 1  # Primeira coluna
mask[:, -1] = 1  # Última coluna

# Aplicar a máscara para destacar as bordas
array[mask == 1] = 255

# Visualizar o array como imagem
plt.imshow(array, cmap='gray')
plt.show()

# Salvar a imagem
plt.imsave('imagem_com_bordas.png', array, cmap='gray')
```

```python
import altair as alt
from vega_datasets import data

source = data.seattle_weather()

alt.Chart(source, title="Daily Max Temperatures (C) in Seattle, WA").mark_rect().encode(
    alt.X("date(date):O").title("Day").axis(format="%e", labelAngle=0),
    alt.Y("month(date):O").title("Month"),
    alt.Color("max(temp_max)").title(None),
    tooltip=[
        alt.Tooltip("monthdate(date)", title="Date"),
        alt.Tooltip("max(temp_max)", title="Max Temp"),
    ],
).configure_view(
    step=13,
    strokeWidth=0
).configure_axis(
    domain=False
)
```

```python
df_acidentes_geral_por_uf_10 = pd.read_csv('datasets/gerais/acidentes_geral_por_uf_10.csv', sep=',', encoding="UTF-8") 
df_acidentes_geral_por_uf_10

filtrado = df_acidentes_geral_por_uf_10[(df_acidentes_geral_por_uf_10['ano'] == 2024)]
filtrado
```

```python
df_acidentes_geral_por_uf_10 = pd.read_csv('datasets/gerais/acidentes_geral_por_uf_10.csv', sep=',', encoding="UTF-8") 
df_acidentes_geral_por_uf_10

#uf,qtd,ano

grafico = alt.Chart(filtrado, title="Mapa de Calor - UF").mark_rect().encode(
    alt.X("ano:Q").title("Ano").axis(format="%e", labelAngle=0),
    alt.Y("uf:N").title("UF"),
    alt.Color("max(qtd)").title(None),
    tooltip=[
        alt.Tooltip("ano", title="Ano"),
        alt.Tooltip("max(qtd)", title="Qtd de Acidentes"),
    ],
).configure_view(
    step=13,
    strokeWidth=0
).configure_axis(
    domain=False
)
grafico
```

```python
# Cria o gráfico de calor
heatmap = alt.Chart(df_acidentes_geral_por_uf_10).mark_rect().encode(
    x=alt.X('ano:O', title=None, axis=alt.Axis(orient='top')),  # Define a orientação do eixo x como 'top'
    y=alt.Y('uf:N', title=None),
    color=alt.Color('qtd:Q', scale=alt.Scale(scheme='yellowgreenblue')),    
)

# Adiciona o texto dentro de cada célula com o valor real e ajusta a cor do texto
text = alt.Chart(df_acidentes_geral_por_uf_10).mark_text(baseline='middle', fontSize=10).encode(
    x='ano:O',
    y='uf:O',
    text=alt.Text('qtd:Q', format='.0f'),
)

# Combina o gráfico de calor com o texto
heatmap_with_text = (heatmap + text).properties(
    width=800,
    height=600,
    title=f'Mapa de Calor por Classificação x Envolvidos (2024)'
)

heatmap_with_text
```

```python
from vega_datasets import data

cars = data.cars.url

selection = alt.selection_point(fields=['Origin'])
color = alt.condition(
    selection,
    alt.Color('Origin:N').legend(None),
    alt.value('lightgray')
)

scatter = alt.Chart(cars).mark_point().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q',
    color=color,
    tooltip='Name:N'
)

legend = alt.Chart(cars).mark_point().encode(
    alt.Y('Origin:N').axis(orient='right'),
    color=color
).add_params(
    selection
)

scatter | legend
```
