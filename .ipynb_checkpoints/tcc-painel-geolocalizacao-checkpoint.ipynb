```python
# ==========================================================
def processa_coordenadas_ano(ano, df):

    
    # Definindo as colunas
    colunas = ['data_inversa', 'uf','municipio', 'br', 'km', 'latitude', 'longitude', 'acidentes']             
    
    # Criando o DataFrame vazio
    df_retorno = pd.DataFrame(columns=colunas)

    lista_latitude = df['latitude'].unique()
    
    print(f'\n\nProcessando o ano de {ano}...\nQtd de latitudes => {len(lista_latitude)}') 
    lista_registros = []
    
    # Colunas ['data_inversa', 'uf','municipio', 'br', 'km', 'latitude', 'longitude','acidentes']
    for indice, item_latitude in enumerate(lista_latitude):
        if indice % 1000 == 0:
            print(f'Processando 1000 registros a partir de {indice}...')
        
        df_filtrado = df[(df['latitude'] == item_latitude)]
        if indice % 1000 == 0:
            print(f'Analisando o item da lista = {item_latitude} | qtd de registros {df_filtrado.shape[0]}...')
        reg = df_filtrado.iloc[0]  

        novo_registro = {
          'data_inversa': reg['data_inversa'], 
          'uf': reg['uf'],
          'municipio': reg['municipio'],
          'br': reg['br'],
          'km': reg['km'],
          'latitude': reg['latitude'],
          'longitude': reg['longitude'],
          'acidentes': df_filtrado.shape[0]
        }       
        
        df_retorno = pd.concat([df_retorno, pd.DataFrame([novo_registro])], ignore_index=True)                   

    df_retorno['latitude'] = df_retorno['latitude'].str.replace(',', '.')
    df_retorno['longitude'] = df_retorno['longitude'].str.replace(',', '.')
    
    return df_retorno           
# ==========================================================
def salva_arquivo_coordenadas_processadas_ano(ano, df):

    arquivo = f'geo/acidentes_localizacao_processado_{str(ano)}.csv'
    df.to_csv(arquivo, sep=';')
    print(f'\n\nSalvando arquivo {arquivo}...')    
# ==========================================================
```

```python
df_coords_2023 = processa_coordenadas_ano('2023', df_acidentes_geo_2023)
salva_arquivo_coordenadas_processadas_ano('2023', df_coords_2023)
```
